```python
# Function to reorder elements of arr[] according 
# to index[] 
def reorder(arr, index, n): 

    # Fix all elements one by one 
    for i in range(0,n): 
  
           #  While index[i] and arr[i] are not fixed 
           while (index[i] != i): 
          
               # Store values of the target (or correct)  
               # position before placing arr[i] there 
                oldTargetI = index[index[i]] 
                oldTargetE = arr[index[i]] 
  
               # Place arr[i] at its target (or correct) 
               # position. Also copy corrected index for 
               # new position 
                arr[index[i]] = arr[i] 
                index[index[i]] = index[i] 
  
               # Copy old target values to arr[i] and 
               # index[i] 
                index[i] = oldTargetI 
                arr[i] = oldTargetE 
```

```python
# Dataframe obtain from google collab
Solution =  pd.read_pickle("Manual_subset.pkl")
```

```python
Solution.reset_index(drop=True,inplace=True)
Solution.drop(columns=['character', 'Text',"rowIndex","Emo_Bert","plutchik_basic"], inplace =True)
#Check if their is nan value
Solution.columns[Solution.isnull().any()].tolist()
```

```python
# Re-order Emobert
Solution.apply(lambda x : reorder(x.Emo_bert_distribution, x.Emo_bert_ids, 41), axis = 1 )
```

```python
# Split Emobert_distribution
df1 = pd.DataFrame(Solution['Emo_bert_distribution'].values.tolist(),columns=labels )
```

```python
# Concatenate
Solution = pd.concat([Solution[:], df1[:]], axis=1)
```

```python
# Drop useless Column
Solution.drop(columns=['Emo_bert_distribution', 'Emo_bert_ids'], inplace =True)
```

```python
Solution["Nbr_Row"] = 1
```

```python
Solution.head(-1)
```

```python
print("Is there dialogs with only one row ?  : ",~Solution['Num_dialog'].duplicated(keep = False).any())
```

```python
result = Solution.groupby(["Num_dialog"]).mean().copy()
```

```python
result["Nbr_Row"] = Solution.groupby(["Num_dialog"]).Nbr_Row.sum().copy()
```

```python
result["Prominent_score"] = result[labels].max(axis=1)
result["Prominent_EmoBert"] =result[labels].idxmax(axis=1)
# Get prominent plutchik 
result["Prominent_Plutchik"] = result.Prominent_EmoBert.progress_apply(lambda x : plutchik[labels.tolist().index(x)])
#Neutrality score
result["Neutrality_score"] = result[neutral].sum(axis=1)
#Emotionality score
result["Emotionality_score"] = result[Emotional].sum(axis=1)
```

```python
result.drop(columns=labels, inplace =True)
```
